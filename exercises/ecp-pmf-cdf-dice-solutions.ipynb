# **EXPLORING PMF AND CDF OF A DISCRETE DISTRIBUTION**

You already know that probability distribution of a discrete random variable $X$can be defined with a probability mass function (PMF) $P_X(x) = P(X = x)$ or, alternatively, using a cumulative disrtibution function (CDF) $F_X(x) = P_X(X 
\leq x)$.

In this notebook, you will consider a random variable $X$ representing the sum of the values obtained after rolling several dice.

Your task is to construct and plot both PMF and CDF of $X$. Follow the steps described below to do so.

## **Rolling two fair dice**

To begin with, let's consider the case of 2 fair dice. Random variable $X$ represents the sum of the values obtained.

What is the range of $X$, and in how many ways can $X$ take each of its values? Count by hand, or write and run a  piece of code that would go over all possible combinations of values on the two dice and count in how many ways $X$ can take each of the values.

```python
counts = {}

for die_1 in [1, 2, 3, 4, 5, 6]:
  for die_2 in [1, 2, 3, 4, 5, 6]:
    s = die_1 + die_2
    if (s in counts.keys()):
      counts[s] += 1
    else:
      counts[s] = 1
```

Now, convert the counts into *probabilities* $P(X = x)$ by dividing the raw counts by the total number of possible outcomes ($6\cdot6 = 36$ for the case of two dice).

```python
def counts_to_proba(counts):

  range = list(counts.keys())

  n = sum(counts.values())
  probabilities = [p/n for p in counts.values()]

  return range, probabilities

range_2, probabilities_2 = counts_to_proba(counts)
```

Note that probabilities $P(X=x)$ for every possible $x$ frm the range $R_X$ of $X$ fully describe the probability mass function $P_X(x)$.

Plot it. **Don't forget that $P_X(x)$ is only defined at the points $x \in R_X$, so the plot shouldn't look like a single line.**

```python
plt.plot(range_2, probabilities_2, 'o', label='$P_X(x)$')
plt.plot(range_2, probabilities_2, 'grey')
_ = plt.xticks(range(0, max(range_2)+1))
_ = plt.title('Rolling two fair dice, $X$ - sum of outcomes')
_ = plt.xlabel('$x$')
_ = plt.ylabel('$P(x)$')
plt.legend()
```

Now, construct the CDF $F_X(x) = P_X(X \leq x)$ from the PMF $P_X(x) = P(X=x)$ obtained previously.

*Tip: function `cumsum()` from the `numpy` library might come in handy.*

```python
def probabilities_to_cdf(probabilities):
  cdf = np.cumsum(probabilities)
  return cdf

cdf_2 = probabilities_to_cdf(probabilities_2)
```

Finally, make a plot for the CDF $F_X(x) = P(X \leq x)$. 

Don't forget that CDF $F_X(x)$ is a step function that has a 'jump' at each value from range $R_X$ of $X$.

*Tip: function `step()` from the `matplotlip.pyplot` library can be handy for plottin such a step function.*

```python
plt.step(x=range_2, y=cdf_2, where='post', label='CDF $F_X(x)')
_ = plt.xticks(range(0, max(range_2)+1))
_ = plt.title('Rolling two fair dice, $X$ - sum of outcomes')
_ = plt.xlabel('$x$')
_ = plt.ylabel('$P(X \leq x)$')
_ = plt.legend()
```

## **Rolling three fair dice**

Now, consider rolling three fair dice instead of just two. Random variable $X$ still represents the sum of the obtained values.

Repeat the steps above in order to construct and plot both PMF $P_X(x) = P(X=x)$ and CDF $F_X(x) = P(X \leq x)$ in this case.

```python
counts = {}

for die_1 in [1, 2, 3, 4, 5, 6]:
  for die_2 in [1, 2, 3, 4, 5, 6]:
    for die_3 in [1, 2, 3, 4, 5, 6]:
      s = die_1 + die_2 + die_3
      if (s in counts.keys()):
        counts[s] += 1
      else:
        counts[s] = 1
```

```python
range_3, probabilities_3 = counts_to_proba(counts)
```

```python
plt.plot(range_3, probabilities_3, 'o', label='$P_X(x)$')
plt.plot(range_3, probabilities_3, 'grey')
_ = plt.xticks(range(0, max(range_3)+1))
_ = plt.title('Rolling three fair dice')
_ = plt.xlabel('$x$')
_ = plt.ylabel('$P(x)$')
_ = plt.legend()
```

```python
cdf_3 = probabilities_to_cdf(probabilities_3)

plt.step(x=range_3, y=cdf_3, where='post', label='$F_X(x)$')
_ = plt.xticks(range(0, max(range_3)+1))
_ = plt.title('Rolling three fair dice, $X$ - sum of outcomes')
_ = plt.xlabel('$x$')
_ = plt.ylabel('$P(X \leq x)$')
_ = plt.legend()
```

```python
counts = {}

for die_1 in [1, 2, 3, 4, 5, 6]:
  for die_2 in [1, 2, 3, 4, 5, 6]:
    for die_3 in [1, 2, 3, 4, 5, 6]:
      for die_4 in [1, 2, 3, 4, 5, 6]:
        for die_5 in [1, 2, 3, 4, 5, 6]:
          s = die_1 + die_2 + die_3 + die_4 + die_5
          if (s in counts.keys()):
            counts[s] += 1
          else:
            counts[s] = 1

range_5, probabilities_5 = counts_to_proba(counts)  
cdf_5 = probabilities_to_cdf(probabilities_5)
```
