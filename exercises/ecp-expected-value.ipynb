```python
import random
```

```python
N = 100
p = 0.8

# 1. Toss a coin N times (1 = H, 0 = T)
# 2. Count the number of H

def toss_coin(p):
  if (random.random() < p):
    return 1
  else:
    return 0

tosses = [toss_coin(p) for _ in range(N)]
y = sum(tosses)
y
```

```python
def generate_number():
  # P(-1) = P(0) = 0.1
  # P(1) = 0.8

  r = random.random()

  if (r<0.1):
    return -1
  elif (r<0.2):
    return 0
  else:
    return 1

N = 10000000
numbers = [generate_number() for _ in range(N)]
print(numbers[:20])
print(sum(numbers)/N)
```
